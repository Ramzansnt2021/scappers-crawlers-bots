```python
#################################\
# Industrial Network Infrastructure/Secure Routers/Secure Routers/EDR-810 Series
source = "D://UAE//moxa-eworldme//"
main_category = []
categories = []
Brands = []
sub_cate_1 = []
sub_cate_2 = []
sub_cate_3 = []
sub_cate_4 = []
currency = []
prices = []
SKUS = []
descriptions = []
short_descriptions = []
product_images = []
product_prices = []
prod_model_num = []
listToStr = Taggs = []

option = Options()
option.add_argument("--disable-infobars")
option.add_argument("start-maximized")
option.add_argument("--disable-extensions")

# Pass the argument 1 to allow and 2 to block
option.add_experimental_option("prefs", { 
    "profile.default_content_setting_values.notifications": 1 
})
option.headless = True
driver = webdriver.Chrome(chrome_options=option, executable_path="D://chromedriver")

# link = "https://moxa.eworldme.com/MOXA-Industrial-Edge-Connectivity/MOXA-USB-to-Serial-Converters-USB-Hubs/MOXA-USB-to-Serial-Converters/MOXA-UPort-2210-2410-Series/MOXA-Uport-2410-USB-to-Serial-Converter"

with open("D://UAE//moxa-eworldme//upload_files//IndustrialEdgeConnectivity_2.txt", 'r', encoding= 'utf-8') as fp:
    url = fp.readlines()
    for link in url:
        driver.get(link)
        driver.implicitly_wait(10)
        time.sleep(10)
        page = driver.find_elements_by_class_name('site-wrapper')
        for pages in page:
            soup = BeautifulSoup(pages.get_attribute('innerHTML'), 'html.parser') 
            breadcrumbs = soup.select('.breadcrumb')
            for sub_list in breadcrumbs:
                c_list = sub_list.text
                categories.append(c_list.strip().replace('\n',' '))
        details = soup.select('#product,.product-details')
        for descript in details:
            try:
                short = descript.select('.title,.page-title')
                for sh in short:
                    descriptions.append(sh.text)
            except Exception as e:
                sh = None
        for p in details:
            try:
                price = p.select('.product-price')
                for pr in price:
                    product_prices.append(pr.text)
            except Exception as e:
                p = None
        for m in details:
            try:
                model = m.select('.product-model')
                for m_no in model:
                    prod_model_num.append(m_no.text)
            except Exception as e:
                m = None
        img = soup.select('.product-image, .direction-vertical, .position-left')
        try:
            for i in img:
                g = i.select('swiper-slide, .swiper-slide-visible')
                for images in g:
                    link = images.find('img')['src']
                    product_images.append(link)
        except Exception as e:
            img = None
    fp.close()



for sku in prod_model_num:
    PID = sku.split()[1:]
    listToStr = ' '.join(map(str, PID)) 
    SKUS.append(listToStr)
# BRNDS
for brand in categories:
    br = brand.split()
    
    br = brand.split()[:3]
    
    cat = brand.split()[3:6]
    
    
    cateOne = brand.split()[6:8]
    
    
    cateTwo = brand.split()[8:11]

    
    cateThri = brand.split()[12:]
    
    
#     using list comprehension 
    brandlist = ' '.join(map(str, br)) 
    Brands.append(brandlist)

#     using list comprehension 
    catlist = ' '.join(map(str, cat)) 
    main_category.append(catlist)

#     using list comprehension 
    oneCate = ' '.join(map(str, cateOne)) 
    sub_cate_1.append(oneCate)
    
#     using list comprehension 
    twoCate = ' '.join(map(str, cateTwo)) 
    sub_cate_2.append(twoCate)
    
#     using list comprehension 
    thriCate = ' '.join(map(str, cateThri)) 
    sub_cate_3.append(thriCate )
    

# Currrency
for price in product_prices:
    pr = price.split()[0]
    currency.append(pr)

# Price Value
for price in product_prices:
    pr = price.split()[1]
    prices.append(pr)

# Short description
for sh in descriptions:
    des = sh.split()[1:]
    descript = ' '.join(map(str, des)) 
    short_descriptions.append(descript)
    
data = {'SKUs':SKUS, 'product_name':prod_model_num,'Brand':Brands, 'Main Category':main_category, 'Breadcrumbs':categories,
        'Sub Category 1': sub_cate_1, 'Sub Category 2': sub_cate_2, 'Sub Category 3': sub_cate_3, 
        'Short Description':short_descriptions, 'Product Images':product_images, 'Products Prices':product_prices, 
        'Product Currency': currency, 'Product Prices':prices,}
# print("\n#########\n",data)

df = pd.DataFrame.from_dict(data, orient='index')
df = df.transpose()
print(df)
print("Got these many results:",df.shape)
    
df.to_csv(source+""+"IndustrialEdgeConnectivity_2.csv",index=False)
```
